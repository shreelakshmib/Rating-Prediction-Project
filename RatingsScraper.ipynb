```python
import time
import pandas as pd
from selenium import webdriver
```

```python
star_list = []
review_list = []

ecommerce_Websites = ['https://www.amazon.in/','https://www.flipkart.com/']
search_objects = ['laptops', 'mobile phones', 'headphones','smart watches','professional cameras','printers',
                  'home theater', 'router', 'mouse','pendrive','speaker','hard disk','smart bands','monitors','ipad',
                  'security camera','internal ssd', 'ram', 'Tablets','Memory Cards','webcam','keyboards','power bank','projectors']
selector = {
    'https://www.amazon.in/':     {
                                    'product_url' : '//a[@class="a-link-normal a-text-normal"]',
                                    'star' : '//i[contains(@data-hook,"review-star-rating")]',
                                    'reviews' : '//div[@data-hook="review-collapsed"]',
                                    'pages' : "//*[@id='search']/div[1]/div/div[1]/div/span[3]/div[2]/div[21]/span/div/div/ul/li/a"
                                  },
    'https://www.flipkart.com/' : {
                                    'product_url' : '//*[@id="container"]/div/div[3]/div[1]/div[2]/div/div/div/div/a',
                                    'star' : '//div[@class="_2c2kV-"]/div/div/div/div[1]/div',
                                    'reviews' : '//div[@class="_2c2kV-"]/div/div/div/div[2]',
                                    'pages': "//a[contains(@class,'ge-49M')]"
                                  }
}
```

```python
def get_reviews(productList_urls,star_selector, review_selector):
    
    
    for each in productList_urls:
        driver.get(each)
        print(each)

        try:
            star_tags = driver.find_elements_by_xpath(star_selector)
            #print(len(star_tags))
            for star in star_tags:
                if star is None:
                     star_list.append("N/A")
                else:
                    product_stars = star.get_attribute('innerText').replace(".0 out of 5 stars","")
                    star_list.append(product_stars) 
        except Exception as err:
            print(err)

        print("Got the Stars : ", len(star_list))


        try:
            #print("Trying to retrieve reviews : ")
            review_tags = driver.find_elements_by_xpath(review_selector)
            #print(len(review_tags))
            for review in review_tags:
                if review.text is None:
                    review_list.append("NA")
                else:
                    product_review = review.get_attribute('innerText').replace("\n","").replace("<br>","")
                    review_list.append(product_review)
        except Exception as err:
            print(err)

        print("Got the Reviews : ", len(review_list))
        time.sleep(0.25)
    return star_list,review_list

    
def get_url(url,url_selector,objects=search_objects):
    #print(url,url_selector)
    
    print("Collecting urls from : ", url,"\n")
    
    #driver.get(url)
    #time.sleep(0.25)
    base_url = url
    productList_urls = []

    for objects in search_objects:
        
        driver.get(base_url)
        search_bar = driver.find_element_by_xpath("//input[@type='text']")
        search_bar.send_keys(objects)
        search_bar.submit()
        #search_bar.clear()

        current_url = driver.current_url
        driver.get(current_url)
        time.sleep(0.2)   

            
        url_tags = driver.find_elements_by_xpath(url_selector)
        for url in url_tags:
            href = url.get_attribute('href')
            productList_urls.append(href)
        
        print("\nFound ",len(productList_urls), " urls for ", objects, ".")
        
    
    return productList_urls
    
```

```python
driver = webdriver.Chrome("chromedriver.exe")

for websites in ecommerce_Websites:
    
    for i,x in enumerate(selector):
        if (x == websites):
            url_selector = selector[websites]['product_url']
            star_selector = selector[websites]['star']
            review_selector = selector[websites]['reviews']
    
    url_list = get_url(websites,url_selector, search_objects)
    ratings_list, reviews_list = get_reviews(url_list,star_selector, review_selector)
    

driver.close()

final_df = pd.DataFrame({'stars': ratings_list,
                     'reviews' : reviews_list})
final_df.to_csv("Product Reviews.csv")
```
